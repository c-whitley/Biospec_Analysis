```python
from skimage.filters import threshold_otsu\
, threshold_triangle, threshold_mean\
, sobel, threshold_li, threshold_minimum\
, threshold_isodata, threshold_adaptive

from skimage.segmentation import clear_border
from skimage.measure import label, regionprops
from skimage.morphology import closing, square
from skimage.color import label2rgb
import matplotlib.patches as mpatches

image = amide_1

# apply threshold
thresh = threshold_isodata(amide_1)
bw = closing(image > thresh, square(7))

block_size = 2801
bw = threshold_adaptive(amide_1, block_size, offset = 0)
bw = closing(bw, square(3))

# remove artifacts connected to image border
cleared = clear_border(bw)

# label image regions
label_image = label(cleared)
image_label_overlay = label2rgb(label_image, image=image)


fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(image_label_overlay)


regions = regionprops(label_image)


for region in regionprops(label_image):
    
    # take regions with large enough areas which are roughly circular
    if region.area >= 5000 and region.eccentricity < 0.8:
        #if ((region.bbox[0] - region.bbox[1])/(region.bbox[2] - region.bbox[3])) > 1.5:
        # draw rectangle around segmented coins
        minr, minc, maxr, maxc = region.bbox
        rect = mpatches.Rectangle((minc, minr), maxc-minc, maxr-minr,
                                  fill=False, edgecolor='red', linewidth=2)
        ax.add_patch(rect)
        ax.text(minc, minr, region.label, color = "w")

#[plt.axvline(x) for x in np.linspace(100,3200,10)]        

plt.tight_layout()
plt.show()
```

```python
from tqdm import tqdm

biggest = np.argmax([region.area for region in regions])
biggest_shape = regions[biggest].image.shape

masks = {name: imageio.imread(r"C:\Users\Conor\Documents\Breast_Tissue_Labels\2085-b\masks\2085-b_{}_labels.png".format(name)) for name in tissue_colours.values()}

for i, region in enumerate(tqdm([image for image in regionprops(label_image) if image.area >= 5000 and image.eccentricity < 0.8])):
    
    if region.label in [938, 917, 464]: continue
    
    #if i not in [14]: continue
    
    bbox = region.bbox # Get bounding box of the segmented image
    
    temp_im = img.read_subregion([bbox[0], bbox[2]], [3200+bbox[1], 3200+bbox[3]]).mean(axis = 2) # Get the cropped datacube
    temp_im = gray2rgb(temp_im)
    
    label_img = np.zeros((temp_im.shape[0], temp_im.shape[1], 3), dtype = np.uint8)

    for colour, name in tissue_colours.items():
        
        labels = masks[name]
        
        cropped_labels = labels[bbox[0]:bbox[2], bbox[1]: bbox[3]]
        
        # Apply mask to blank image to get colour
        temp_im[cropped_labels > 1] = np.array(colour)
        
        #scipy.misc.imsave(r"C:\Users\Conor\Documents\Breast_Tissue_Labels\{}\{}.jpg".format(name, region.label), label_img)
        
    #if label_img.sum() == 0: continue # Don't bother saving blank images
    #if i > 10: break
        
    imageio.imsave(r"C:\Users\Conor\Documents\Breast_Tissue_Labels\2085-b\labelled_cores\{}.png".format(region.label), temp_im)
```
